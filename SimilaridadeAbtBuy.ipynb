Como será feita uma análise de todos os registros de uma tabela com todos de outra, é conveniente criarmos um índice para aumentarmos a performance do treinamento.

```python
pip install recordlinkage
```

```python
import recordlinkage
indexer = recordlinkage.Index()
indexer.full()
```

```python
combinacoesCandidatas = indexer.index(dataFrameAbt, dataFrameBuy)
```

```python
combinacoesCandidatas
```

Agora que o índice está criado, vamos partir para a comparação dos valores. Primeiro, vamos configurar qual abordagem será utilizada na comparação.

```python
comparador = recordlinkage.Compare()
comparador.string('name', 'name', method="jarowinkler")
```

Agora que o objeto que fará a coparação está criado, vamos comparar os valores para obter o score de similaridade.

```python
features = comparador.compute(combinacoesCandidatas, dataFrameAbt, dataFrameBuy)
```

```python
features
```

Vamos filtrar para exibir os produtos com mais de 0.85 de similaridade.

```python
similares = features[features[0]>0.9]
len(similares)
```

```python
similares.head(15)
```

Agora que vimos que tem 190 produtos similares, vamos ver quais são

Primeiro vamos reiniciar os índices dos dataframes para deixar a consulta mais performatica

```python
dataFrameAbt = dataFrameAbt.reset_index()
dataFrameBuy = dataFrameBuy.reset_index()
```

Vamos avaliar os nomes e as descrições dos registros com correspondencia maior que 90%

```python
paresSimilares = []
for index_abt, index_buy in similares.index:
  record_abt = dataFrameAbt.loc[index_abt]
  record_buy = dataFrameBuy.loc[index_buy]
  paresSimilares.append((record_abt, record_buy))
```

Agora que foram recuperadas as correspondencias, vamos ver os registros

```python
for i, (record_abt, record_buy) in enumerate(paresSimilares):
  print("Par:",i)
  print("Registro Abt:")
  print(record_abt)
  print("Registro Buy:")
  print(record_buy)
  print("="*100)
```
