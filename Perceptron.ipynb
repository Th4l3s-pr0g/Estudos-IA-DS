Passo 1 -  Definir uma variável para armazenar o tamanho da amostra que será utilizado no treinamento do perceptron.

```python
tamAmostra = 6
print("Tamanho da amostra:", tamAmostra)
```

Passo 2 -  Criar variáveis para armaxenar os valores das frutas da amostra. Vamos iniciar a utilização de uma nova biblioteca, a NumPy. A NumPy é muito eficiente na realização de operações matemáticas que envolvem matrizes/listas. Primeiro iremos importar a biblioteca e depois criaremos as variáveis. 

```python
import numpy as np
massa = np.array([113,122,107,98,115,120])
pH = np.array([6.8,4.7,5.2,3.6,2.9,4.2])
y = np.array([-1,1,-1,-1,1,1])
print("Massa das frutas:", massa)
print("pH das frutas:", pH)
print("Código das frutas:", y)
```

Passo 3 - uma vez que os arrays coms os valores da amostra foram criados, é coerente separarmos as variáveis INDEPENDENTES (X) da DEPENDENTE. As variáveis INDEPENDENTES são massa e pH, logo, esses atributos serão armazenados na variável X. Para armazenarmos as variáveis INDEPENDENTES em X, vamos empilhar o vetores massa e pH.

```python
X = np.vstack((massa,pH))
print(X)
```

Passo 4: uma vaez que separamos as variáveis INDEPENDENTES(massa e pH) da variável DEPENDENTE(fruta), vamos criar uma variável para armazenar o valor do bias.

```python
bias = -1
print(bias)
```

Passo 5: vimos que a fórmula que recalcula os valores dos pesos possui uma taxa de aprendizagem. Vimos também que essa taxa deve ter o valor entre 0 e 1.

```python
txAprendizado = 0.1
print("Taxa de Aprendizado:", txAprendizado)
```

Passo 6: já definimos os valores das entradas (X), doa bias e da taxa de aprendizado. Agora vamos definir os valores iniciais dos pesos (w).

```python
w = np.zeros([3])
print("Pesos:", w)
```

Passo 7: uma vez que todas as variáveis foram estruturadas, vamos criar uma lista que armazenará os erros encontrados durante o treinamento do neurônio.

```python
erros = np.zeros([6])
print("Lista que armazenará os erros:", erros)
```

Passo 8: agora que toda a estrutura de dados foi definida, vamos definir a função de ativação. No nosso caso, temos dois tipos de fruta (laranja e maçã), diante disso utilizaremos a função degrau, pois precisamos retornar dois valores. No nosso caso, se rtornar -1 = maçã; 1 = laranja.

```python
def funcaoAtivacao(u):
  if u < 0.0:
    return(-1) #maçã
  else:
    return(1) #laranja
```

Vamos testar a função de ativação criada.

```python
uTeste = -30
print("Valor resornado pela ativação:", funcaoAtivacao(uTeste))
```

Passo final: definimos tudo que é necessário para treinar um neurônio. Agora partiremos para o treinamento. Lembre-se que o principal objetivo do treinamento é calibrar os valores dos pesos(w).

```python
numEpocas = 63725
for i in range(numEpocas):
  for j in range(tamAmostra):
    #1 - vamos criar uma lista Xb que contemplará os valores da massa, do pH e do bias.
    Xb = np.hstack((bias, X[:,j]))
    #print(Xb)
    #2 - Vamos calcular o valor de u para cada fruta
    u = Xb[0]*w[0] + Xb[1]*w[1] + Xb[2]*w[2]
    #print(u)
    #3 - Vamos pedir para o neurônio mostrar a fruta  proposta a partir da função de ativação
    Yr = funcaoAtivacao(u)
    #print("Fruta prevista pelo neurônio:", Yr)
    #4 - uma vez que Yr mostra a fruta prevista pelo neurônio, vamos ver se ele acertou ou errou
    erros[j] = y[j] - Yr
    #print("Erros:",erros[j])
    if(erros[j] !=0.0):
      w = w + txAprendizado*erros[j]*Xb
  print("Ao final da época", i, "os erros foram", str(erros), "e os pesos", str(w))
  if(erros[0] == 0 and erros[1] == 0 and erros[2] == 0 and erros[3] == 0 and erros[4] == 0 and erros[5] == 0):
    break
```
